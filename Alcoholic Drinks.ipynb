```python
import matplotlib.pyplot as plt
plt.scatter(ads['beer_servings'], ads['total_litres_of_pure_alcohol'], marker='o', color='orange')
plt.xlabel('Beer Servings')
plt.ylabel('Total Litres of Pure Alcohol')
plt.legend(loc='upper right')
plt.show()
```

```python
import seaborn as sns
%matplotlib inline
sns.heatmap(ads.corr(), annot=True, fmt=".2f")
plt.show()
```

```python
df_heatmap = ads.pivot_table(index='country', columns='beer_servings', values='total_litres_of_pure_alcohol', aggfunc=np.mean)
sns.heatmap(df_heatmap, annot=True, fmt=".1f")
plt.show()
```
